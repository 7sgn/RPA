```python
# Comentado com chat GPT
# Importação das bibliotecas necessárias
import pandas as pd  # Para manipulação de dados e exportação para CSV
from selenium import webdriver  # Para controlar o navegador via Selenium
from selenium.webdriver.common.by import By  # Para localizar elementos da página
from selenium.webdriver.chrome.options import Options  # Para configurar opções do Chrome
from fake_useragent import UserAgent  # Para gerar um user-agent aleatório
import random  # Para gerar números aleatórios, usados para o intervalo de tempo
import time  # Para controlar o tempo de espera entre as ações

# Criação de um objeto UserAgent para gerar user-agents aleatórios
ua = UserAgent()

# Configuração das opções do Chrome
chrome_options = Options()
chrome_options.add_argument(f'user-agent={ua.random}')  # Configura o user-agent aleatório

# Inicialização do driver do Chrome com as opções configuradas
driver = webdriver.Chrome(options=chrome_options)

# URL do Mercado Livre que será acessada
url = 'https://www.mercadolivre.com.br/ofertas'
driver.get(url)  # Acessa a página

# Tempo de espera para garantir que a página carregue
driver.implicitly_wait(5)

# Lista para armazenar os dados dos produtos extraídos
product = []

# Loop para extrair informações dos produtos presentes na página
for item in driver.find_elements(By.CLASS_NAME, 'poly-card__content'):
    try:
        # Extrai o nome, preço e link de cada produto
        name = item.find_element(By.CLASS_NAME, 'poly-component__title').text
        price = item.find_element(By.CLASS_NAME, 'poly-component__price').text
        link = item.find_element(By.TAG_NAME, 'a').get_attribute('href')
        
        # Adiciona as informações extraídas à lista 'product'
        product.append({'Nome': name, 'Preço': price, 'Link': link})
    except Exception as e:
        # Caso haja algum erro ao extrair os dados de um produto
        print(f"Erro ao extrair informações: {e}")
        
    # Intervalo aleatório entre as requisições para evitar bloqueios por parte do site
    time.sleep(random.uniform(2, 5))

# Finaliza a sessão do driver
driver.quit()

# Criação de um DataFrame com os dados extraídos
df = pd.DataFrame(product)

# Salvamento dos dados extraídos em um arquivo CSV
df.to_csv('produtos_mercado_livre.csv', index=False, encoding='utf-8')

# Exibe uma mensagem de sucesso
print("Dados extraídos e salvos com sucesso!")





    






```
